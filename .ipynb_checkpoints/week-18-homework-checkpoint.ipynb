## 1.What is a neural network? What are the general steps required to build a neural network? 

Neural network represents the structure of a human brain modeled on the computer. 
It consists of neurons and synapses organized into layers like an input layer,one or more hidden layers and an output layer.

ANN can have millions of neurons connected into one system, which makes it extremely successful at analyzing and 
even memorizing various information.

Neural networks are used to solve complex problems that require analytical calculations similar to those of the human brain. 
The most common uses for neural networks are:Classification,prediction and recognition.
    
Steps required to build a neural network:

    1) Import the data

    2) Split into train and test 
    
    3) Standardize our dataset
    
    4) Define Network(input layer,hidden layer and output layer)
    
    5) Compile Network
    
    6) Evaluate Network
    
    7)Make Predictions
    
Each layer must have an activation function (like ReLU,sigmoid function) and a number of nodes. Models are optimized using a function like Adam or Stochastic Gradient Descent. Neural networks are refined using forward and backward propagation.    
  

## 2.Generally, how do you check the performance of a neural network? Why?

There are different metrics for chrecking the performance of neural network 
Regression metrics :

    Mean squared error 
    
    Mean absolute error
    
    Mean absolute percentage error 
    
    Cosine proximity 
    
    Root mean square error
Classification Metrics :

    Binary accuracy

    categorical accuracy

    sparse categorical accuracy 


We  tune the neural network with loss function. The neural network looks at the target and predicition, calculates the loss, then backpropagates through the network to change the weights.

## Solving it as a Classification problem

```python
#Importing libraries 
import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
#from keras.utils import to_categorical

X = abalone_df[['length','diameter','height','whole_wt','shucked_wt','viscera_wt','shell_wt','sex_cat']]
y = abalone_df['rings_bin']

# Split into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42, stratify=y)


# #Standardize
sc= StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)


from keras.models import Sequential
from keras.layers import Dense


#Defining model
model = Sequential()
model.add(Dense(4, input_dim=8, activation='relu'))
model.add(Dense(4, activation='relu'))
#model.add(Dense(8, activation='relu'))
model.add(Dense(4, activation='softmax'))
          
#Compile model

model.compile(optimizer=keras.optimizers.Adam(),
         loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True), # default from_logits=False
        metrics=[keras.metrics.SparseCategoricalAccuracy()])

#model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy']) 
                             
model.fit(X_train, y_train,epochs=100, batch_size=10, verbose=1)

pred_train= model.predict(X_train)

scores = model.evaluate(X_train, y_train, verbose=0)
print('Accuracy on training data: {}% \n Error on training data: {}'.format(scores[1], 1 - scores[1]))   
 
pred_test= model.predict(X_test)
scores2 = model.evaluate(X_test, y_test, verbose=0)
print('Accuracy on test data: {}% \n Error on test data: {}'.format(scores2[1], 1 - scores2[1])) 
```

## Solving  as a Regression problem 

```python
X = abalone_df[['length','diameter','height','whole_wt','shucked_wt','viscera_wt','shell_wt','sex_cat']]
y = abalone_df['rings']

# Split into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42, stratify=y)


# #Standardize
sc= StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)

from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import RMSprop
model = Sequential()
model.add(Dense(units=10, input_dim=8,activation='relu'))
model.add(Dense(units=1,activation='linear'))
model.compile(optimizer='adam', loss='mean_squared_error',  metrics=['mae','mse'])

#early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)
model.fit(X_train,y_train,batch_size=5,epochs=1000)
# Model summary for number of parameters use in the algorithm
model.summary()
```

```python
#Predict testing labels
y_pred= model.predict(X_test)

#Calculate RMSE and R^2
from sklearn.metrics import mean_squared_error
from math import sqrt
rms = sqrt(mean_squared_error(y_test, y_pred))
print(rms)
from sklearn.metrics import r2_score
r_squared=r2_score(y_test,y_pred)
print(r_squared)
```

By using keras for regression problem  we got r2_score as 0.52 and RMSE as 1.96

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = abalone_df[['length','diameter','height','whole_wt','shucked_wt','viscera_wt','shell_wt','sex_cat']].values
y = abalone_df['rings'].values

# Split into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42) #, stratify=y)

# #Standardize
sc= StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
print(X_test.shape,y_test.shape)
print(X_train.shape,y_train.shape)
```

```python
import torch.nn as nn
import torch.nn.functional as F#this has activation functions
import pandas as pd
import torch

# Creating tensors
X_train = torch.FloatTensor(X_train)
X_test = torch.FloatTensor(X_test)

y_train = torch.FloatTensor(y_train)
y_test = torch.FloatTensor(y_test)

# torch.FloatTensor(y_train.to_numpy())
#y_test = torch.FloatTensor(y_test.to_numpy())
print(X_train)
```

```python
class ANN_Model(nn.Module):
    def __init__(self, input_features=8, hidden1=20, hidden2=20, out_features =1):
        super().__init__()
        self.layer_1_connection = nn.Linear(input_features, hidden1)
        self.layer_2_connection = nn.Linear(hidden1, hidden2)
        self.out = nn.Linear(hidden2, out_features)
    
    def forward(self, x):
        #apply activation functions
        x = F.relu(self.layer_1_connection(x))
        x = F.relu(self.layer_2_connection(x))
        x = self.out(x)
        return x
```

```python
class ANN_Model(nn.Module):
    def __init__(self, input_features=8, hidden1=20, hidden2=20, hidden3=20, out_features =1):
   # def __init__(self, input_features=8, hidden1=20,hidden2=20, out_features =1):
        super().__init__()
        self.layer_1_connection = nn.Linear(input_features, hidden1)
        self.layer_2_connection = nn.Linear(hidden1, hidden2)
        self.layer_3_connection = nn.Linear(hidden2, hidden3)
        self.out = nn.Linear(hidden3, out_features)
    
    def forward(self, x):
        #apply activation functions
        x = F.relu(self.layer_1_connection(x))
        x = F.relu(self.layer_2_connection(x))
        x = F.relu(self.layer_3_connection(x))
        x = self.out(x)
        return x

```

```python
torch.manual_seed(42)

#instantiate the model
model = ANN_Model()
```

```python
# loss function
loss_function = torch.nn.MSELoss()

#optimizer
optimizer = torch.optim.SGD(model.parameters(),lr=0.001)
```

```python
#loss_function = nn.CrossEntropyLoss()
#optimizer
#optimizer = torch.optim.Adam(model.parameters(), lr = 0.01)
#optimizer = torch.optim.SGD(model.parameters(), lr=0.1, momentum=0.9)
    
```

```python
#run model through multiple epochs/iterations
final_loss = []
n_epochs = 500
for epoch in range(n_epochs):
    y_pred = model.forward(X_train)
    loss = loss_function(y_pred, y_train)
    final_loss.append(loss)
    
    if epoch % 10 == 1:
        print(f'Epoch number: {epoch} with loss: {loss.item()}')
    
    optimizer.zero_grad() #zero the gradient before running backwards propagation
    loss.backward() #for backward propagation 
    optimizer.step() #performs one optimization step each epoch
```

```python
#predictions
y_pred = []

with torch.no_grad():
    for i, data in enumerate(X_test):
        prediction = model(data)
        y_pred.append(prediction.item())
```

```python
#Predict testing labels
#y_pred= model.predict(X_test)
#print(y_pred)
#Calculate RMSE and R^2
from sklearn.metrics import mean_squared_error
from math import sqrt
rms = sqrt(mean_squared_error(y_test, y_pred))
print(rms)
from sklearn.metrics import r2_score
r_squared=r2_score(y_test,y_pred)
print(r_squared)
```
